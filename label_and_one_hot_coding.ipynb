LABEL ENCODING

```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
df=pd.read_csv("customer_data_for_ml_practice.csv")
df2=df.copy()
label=LabelEncoder()
df2['Gender_Encoded']=label.fit_transform(df['Gender'])

```

```python
df2
```

CLEANING DATA

```python
age=df2['Age']
def age_group(age):
    if age <= 12:
        return "Child"
    elif age <= 19:
        return "Teen"
    elif age <= 35:
        return "Young Adult"
    elif age <= 60:
        return "Adult"
    else:
        return "Senior"
df2["Age_Group"] = df["Age"].apply(age_group)
```

```python
df2.isnull().sum()
```

```python
df2['Age']=df2['Age'].fillna(df['Age'].median())
df2['Income']=df2['Income'].fillna(df['Income'].median())
```

```python
df2.duplicated().sum()
```

```python


df2.drop_duplicates(inplace=True)
```
