```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
```

```python
data={
    "customer":['aman','riya','ahemed','ali','zara'],
    'age':[20,21,22,23,24],
    'spending':[10000,12000,30000,13000,25000]
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
x=df[['age','spending']]
```

```python
model=KMeans(n_clusters=2,random_state=42,n_init=10)
```

```python
df['Group']=model.fit_predict(x)
plt.figure(figsize=(6,5))
for i in df['Group'].unique():
    group_data=df[df['Group']==i]
    plt.scatter(group_data['age'],group_data['spending'])
    
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
data2={
    'age':[20,21,22,23,24],
    'income':[10000,12000,30000,13000,25000],
    'spending':[30,40,50,40,70],
    'saving':[1000,1500,3000,1200,2000]
}
df=pd.DataFrame(data2)
scale=StandardScaler()
scaled_data=scale.fit_transform(df)
pca=PCA(n_components=2)
pca_result=pca.fit_transform(scaled_data)
pca_df=pd.DataFrame(pca_result,columns=['pca1','pca2'])
explained_varience=pca.explained_variance_ratio_
print(np.round(explained_varience *100 ,2))
plt.scatter(pca_df['pca1'],pca_df['pca2'])
plt.show()
print(pca_df)
```
